```python
import numpy as np
import nbimporter
from compute_prediction_n_features import compute_prediction_n_features


def compute_cost_n_features(X,y,w,b):
    f_wb = compute_prediction_n_features(X,w,b)
    print(f_wb)
    err = f_wb - y
    cost = (err)**2
    return np.sum(cost)

def compute_cost_n_features_unvec(X,y,w,b):
    m = X.shape[0]
    
    cost = 0
    
    for i in range(m):
        f_wb_i = np.dot(w,X[i]) + b 
        err = f_wb_i - y[i]
        cost += (err) ** 2
        
    return cost
        
```

```python
arr = np.arange(1,21).reshape(5,4)
arr
print(arr[:,1])
```

```python
cost_vec = compute_cost_n_features(arr,arr[:,1],arr[0],10)
cost_unvec = compute_cost_n_features_unvec(arr,arr[:,1],arr[0],10)

print(f"{cost_unvec == cost_vec}")
```
