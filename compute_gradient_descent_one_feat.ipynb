```python
def compute_gradient_one_feature(X,y,w,b):
    m = X.shape[0]
    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):
        f_wb_i = w * X[i] + b
        err = f_wb_i - y[i]
        dj_dw += err * X[i]
        dj_db += err
        
    dj_dw /= m
    dj_db /= m
    
    return dj_dw,dj_db

        
        
```
