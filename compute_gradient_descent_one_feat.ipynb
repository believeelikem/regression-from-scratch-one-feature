```python
def compute_gradient_one_feature(X,y,w,b):
    m = X.shape[0]
    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):
        f_wb_i = w * X[i] + b
        err = f_wb_i - y[i]
        dj_dw += err * X[i]
        dj_db += err
        
    dj_dw /= m
    dj_db /= m
    
    return (dj_dw,dj_db)

        
        
```

```python
import numpy as np 
def compute_gradient_one_feature_vec(X,y,w,b):
    
    m = X.shape[0]
    dj_dw = 0
    dj_db = 0
    
    f_wb = X * w + b
    dj_dw = np.sum((f_wb - y) * X)
    dj_db = np.sum(f_wb - y)
    
    dj_dw /= m
    dj_db /= m
    
    return (dj_dw,dj_db)
```

```python
import pandas as pd

import pandas as pd
import numpy as np


df = pd.read_csv("train.csv")
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['y']).reset_index(drop=True)

x_train,y_train = df["x"].to_numpy(),df["y"].to_numpy()
```

```python
gd_vec = compute_gradient_one_feature_vec(x_train,y_train,100,100)
gd_unvec = compute_gradient_one_feature(x_train,y_train,100,100)

print(gd_unvec)
print(gd_vec)
```

```python
import time

start = time.perf_counter()
vec = compute_gradient_one_feature_vec(x_train,y_train,100,100)
end1 = time.perf_counter()
print(end1-start)

start = time.perf_counter()
unvec = compute_gradient_one_feature(x_train,y_train,100,100)
end2 = time.perf_counter()
print(end2-start)

print(end1 > end2)
```
