```python
def compute_cost_one_feature(X,y,w,b):
    
    m = X.shape[0]
    cost = 0
    
    for i in range(m):
        f_wb = w * X[i] + b
        err = (f_wb - y[i])**2
        cost+=err
        
    total_cost = 1/(2 * m) * cost
    
    return total_cost
```

### OR VECTORISED

```python

import numpy as np

def compute_cost_one_feature_vec(X,y,w,b):
    
    cost = 0
    
    f_wb = X * w + b
    err = (f_wb - y ) ** 2
    cost = np.sum(err)
    
    return cost
```
