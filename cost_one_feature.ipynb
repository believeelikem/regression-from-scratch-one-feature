```python
def compute_cost_one_feature(X,y,w,b):
    
    m = X.shape[0]
    cost = 0
    
    for i in range(m):
        f_wb = w * X[i] + b
        err = (f_wb - y[i])**2
        cost+=err
        
    total_cost = 1/(2 * m) * cost
    
    return total_cost
```

### OR VECTORISED

```python

import numpy as np

def compute_cost_one_feature_vec(X,y,w,b):
    
    cost = 0
    m = X.shape [0]
    
    f_wb = X * w + b
    err = (f_wb - y ) ** 2
    cost = np.sum(err)
    total_cost = 1/(2 * m) * cost
    
    return total_cost
```

```python
import pandas as pd
import numpy as np


df = pd.read_csv("train.csv")
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['y']).reset_index(drop=True)
x_train,y_train = df["x"].to_numpy(),df["y"].to_numpy()
```

```python
import time

start = time.perf_counter()
vec = compute_cost_one_feature_vec(x_train,y_train,100,100)
end1 = time.perf_counter()
print(end1-start)

start = time.perf_counter()
unvec = compute_cost_one_feature(x_train,y_train,100,100)
end2 = time.perf_counter()
print(end2-start)

print(end1 > end2)

```
