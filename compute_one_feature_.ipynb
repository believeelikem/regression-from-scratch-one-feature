### Implementing compute_prediction 

```python
def compute_prediction_one_feature(X,w,b):
    m = len(X)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] += X[i] * w + b
    return f_wb
```

### OR Vectorised

```python
def compute_prediction_one_feature_vec(X,w,b):
    return X * w + b
```
